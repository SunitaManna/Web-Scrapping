```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.amazon.in/s?k=bags+for+women&crid=1ERAUEEY81M92&sprefix=bags%2Caps%2C178&ref=nb_sb_ss_ts-doa-p_2_4'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
title=soup.title
```

```python
print(title.string)
```

```python
Price=soup.find_all('span',class_='a-price-whole')
```

```python
Price
```

```python
Bagprice=[]
for i in range(0,len(Price)):
    Bagprice.append(Price[i].get_text())
```

```python
Bagprice
```

```python
Name=soup.find_all('span',class_='a-size-base-plus a-color-base')
Name
```

```python
Bagname=[]
for i in range(0,len(Name)):
    Bagname.append(Name[i].get_text())
```

```python
Bagname
```

```python
Description=soup.find_all('span',class_='a-size-base-plus a-color-base a-text-normal')
Description
```

```python
BagDescription=[]
for i in range(len(Description)):
    BagDescription.append(Description[i].get_text())
```

```python
BagDescription
```

```python
df=pd.DataFrame()
df["Price"]=Bagprice
df["Name"]=Bagname
df["Description"]=BagDescription
```

```python
df
```

```python
def export_data(data):
    df=pd.DataFrame(data)
    df.to_excel("Amazon Bags.xlsx")
```

```python
export_data(df)
```
